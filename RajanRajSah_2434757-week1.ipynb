Display Fibonacci series up to 10 terms
ANS:Start with 0, 1, then each next term is previous + current.

```python
# Fibonacci series up to 10 terms
a, b = 0, 1
print("Fibonacci (10 terms):")
for _ in range(10):
    print(a, end=" ")
    a, b = b, a + b
print()
```

Find the factorial of a given number (ask user)
Ans: factorial(n) = 1 * 2 * 3 * ... * n (for n >= 0).

```python
# Factorial of a number (user input)
n = int(input("Enter a non-negative integer: "))

if n < 0:
    print("Factorial is not defined for negative numbers.")
else:
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    print(f"Factorial of {n} is {fact}")
```

Print a right angle triangle using for loop
Ans:Print * increasing each row.

```python
# Right angle triangle
rows = int(input("Enter number of rows: "))

for i in range(1, rows + 1):
    print("*" * i)
```

Simple calculator using functions (add, subtract, multiply, divide)

Ans: Define one function per operation.

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: division by zero"
    return a / b

# --- main program ---
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Choose operation: +  -  *  /")
op = input("Enter operation: ").strip()

if op == "+":
    print("Result:", add(num1, num2))
elif op == "-":
    print("Result:", subtract(num1, num2))
elif op == "*":
    print("Result:", multiply(num1, num2))
elif op == "/":
    print("Result:", divide(num1, num2))
else:
    print("Invalid operation.")
```

Remove the first duplicate element of a list

Meaning: Remove the first element that appears again later (i.e., the first value whose count becomes 2 during scanning).

Example: [1, 2, 3, 2, 4] → remove the first duplicated value occurrence → [1, 3, 2, 4] (removes the earlier 2).

```python
def remove_first_duplicate(lst):
    seen = set()
    for i, x in enumerate(lst):
        if x in seen:
            # x is a duplicate; remove its FIRST occurrence (earlier one)
            first_index = lst.index(x)
            lst.pop(first_index)
            return lst
        seen.add(x)
    return lst  # no duplicates found

# Example
my_list = [1, 2, 3, 2, 4, 3]
print("Original:", my_list)
print("After removing first duplicate element:", remove_first_duplicate(my_list.copy()))
```
